### Punto 4. Agregación simple

Calcula:

### Punto 6. JOIN entre tablas

Muestra:
- nombre del estudiante
- nombre del curso
- estado de la inscripción

Usa `estudiantes`, `inscripciones` y `cursos`.

**Entregable:**

```python
# Punto 6 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    e.nombre AS nombre_estudiante,
    c.nombre_curso,
    i.estado
FROM inscripciones i
INNER JOIN estudiantes e
    ON i.estudiante_id = e.estudiante_id
INNER JOIN cursos c
    ON i.curso_id = c.curso_id
"""
run_sql(query)
```

### Punto 7. LEFT JOIN

Trae todos los estudiantes, incluso si no tuvieran inscripciones, mostrando:
- `estudiante_id`
- `nombre`
- `inscripcion_id`

**Entregable:**
- SQL
- pandas

> Pista: en este dataset puede que todos tengan al menos una inscripción, pero igual debes practicar el `LEFT JOIN`.

```python
# Punto 7 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    e.estudiante_id,
    e.nombre,
    i.inscripcion_id
FROM estudiantes e
LEFT JOIN inscripciones i
    ON e.estudiante_id = i.estudiante_id
ORDER BY e.estudiante_id, i.inscripcion_id
"""
run_sql(query)
```

```python
# Punto 7 - pandas
# Escribe aquí tu solución
punto7 = (
    estudiantes
    .merge(inscripciones[["estudiante_id", "inscripcion_id"]],
           on="estudiante_id",
           how="left")
    .sort_values(["estudiante_id", "inscripcion_id"])
)

punto7
```

### Punto 8. HAVING

Calcula cuántas inscripciones tiene cada estudiante y muestra solo aquellos con **2 o más**.

Debes mostrar:
- `estudiante_id`
- `nombre`
- `n_inscripciones`

**Entregable:**
- SQL
- pandas

```python
# Punto 8 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    e.estudiante_id,
    e.nombre,
    COUNT(i.inscripcion_id) AS n_inscripciones
FROM estudiantes e
INNER JOIN inscripciones i
    ON e.estudiante_id = i.estudiante_id
GROUP BY e.estudiante_id, e.nombre
HAVING COUNT(i.inscripcion_id) >= 2
"""
run_sql(query)

```

```python
# Punto 8 - pandas
# Escribe aquí tu solución
punto8 = (
    inscripciones
    .merge(estudiantes, on="estudiante_id", how="inner")
    .groupby(["estudiante_id", "nombre"], as_index=False)
    .agg(n_inscripciones=("inscripcion_id", "count"))
)

punto8 = punto8[punto8["n_inscripciones"] >= 2]

punto8
```

### Punto 9. CASE WHEN / segmentación

Clasifica los cursos según su precio:
- `Bajo` si el precio es menor a 120
- `Medio` si el precio está entre 120 y 180 inclusive
- `Alto` si el precio es mayor a 180

Debes mostrar:
- `nombre_curso`
- `precio`
- `segmento_precio`

**Entregable:**
- SQL
- pandas

```python
# Punto 9 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    nombre_curso,
    precio,
    CASE
        WHEN precio < 120 THEN 'Bajo'
        WHEN precio >= 120 AND 180 THEN 'Medio'
        ELSE 'Alto'
    END AS segmento_precio
FROM cursos
"""
run_sql(query)
```

```python
# Punto 9 - pandas
# Escribe aquí tu solución
import numpy as np

punto9 = cursos[["nombre_curso", "precio"]].copy()

punto9["segmento_precio"] = np.where(
    punto9["precio"] < 120, "Bajo",
    np.where(
        punto9["precio"] <= 180, "Medio",
        "Alto"
    )
)

punto9
```

### Punto 10. Ejercicio final integrador

Calcula el total pagado por categoría de curso.

Debes:
- unir `cursos`, `inscripciones` y `pagos`
- agrupar por `categoria`
- sumar `monto`
- ordenar de mayor a menor

**Entregable:**
- SQL
- pandas

```python
# Punto 10 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    c.categoria,
    SUM(p.monto) AS total_pagado
FROM cursos c
INNER JOIN inscripciones i
    ON c.curso_id = i.curso_id
INNER JOIN pagos p
    ON i.inscripcion_id = p.inscripcion_id
GROUP BY c.categoria
ORDER BY total_pagado DESC
"""
run_sql(query)
```

```python
# Punto 10 - pandas
# Escribe aquí tu solución
punto10 = (
    cursos
    .merge(inscripciones, on="curso_id")
    .merge(pagos, on="inscripcion_id")
    .groupby("categoria", as_index=False)
    .agg(total_pagado=("monto", "sum"))
    .sort_values("total_pagado", ascending=False)
)

punto10
```
