**P2.** Dada la lista `numeros = [10, 25, 3, 47, 8, 15]`:  
- Imprime el primer y el último elemento.  
- Imprime los tres primeros elementos usando slicing.  
- Agrega el número `99` al final de la lista.

```python
numeros = [10, 25, 3, 47, 8, 15]

# Tu respuesta aquí

#Imprime el primer y el último elemento.
primer_numero=numeros[0]
ultimo_numero=numeros[-1]
print(f"El primer número es {primer_numero}")
print(f"El último número es {ultimo_numero}")

#Imprime los tres primeros elementos usando slicing.
tres_primeros_numeros=numeros[:3]
print(f"Los tres primeros numeros son {tres_primeros_numeros}")

#Agrega el número 99 al final de la lista.
numeros.append(99)
print(numeros)
```

**P6.** Crea los siguientes arreglos e imprime su `shape` y `dtype`:  
- Un arreglo con los valores `[1.5, 2.5, 3.5, 4.5]`  
- Una matriz de ceros de 3 filas × 4 columnas  
- Un arreglo con 6 valores igualmente espaciados entre 0 y 1 (usando `linspace`)

```python
# Tu respuesta aquí
#Un arreglo con los valores [1.5, 2.5, 3.5, 4.5]
arreglo=np.array([1.5,2.5,3.5,4.5])
print(arreglo)
print("Shape Arreglo:", arreglo.shape)
print("Dtype Arreglo:", arreglo.dtype)

#Una matriz de ceros de 3 filas × 4 columnas
matriz=np.zeros((3,4))
print("\nMatriz:\n", matriz)
print("Shape Matriz:", matriz.shape)
print("Dtype Matriz:", matriz.dtype)

#Un arreglo con 6 valores igualmente espaciados entre 0 y 1 (usando linspace)
arreglo_linspace=np.linspace(0,1,6)
print("\n",arreglo_linspace)
print("Shape Arreglo Linspace:", arreglo_linspace.shape)
print("Dtype Arreglo Linspace:", arreglo_linspace.dtype)
```

**P7.** Dado el arreglo `a = np.arange(1, 13)`, usa `reshape` para convertirlo en una matriz de 3 × 4.  
Luego imprime el valor máximo, el valor mínimo y la posición (índice) del máximo.

```python
a = np.arange(1, 13)

# Tu respuesta aquí
matriz_2=a.reshape((3,4))
print("\nMatriz:\n", matriz_2)
print("Valor Máximo:", matriz_2.max())
print("Valor Mínimo:", matriz_2.min())
print("Posición del Máximo:", matriz_2.argmax())
```

**P8.** Crea dos arreglos de 5 elementos con `np.random.randint` (valores entre 1 y 20).  
Súmalos, réstalos y multiplícalos elemento por elemento. Imprime cada resultado.

```python
# Tu respuesta aquí
arreglo1=np.random.randint(1,20,5)
arreglo2=np.random.randint(1,20,5)
print("\nArreglo 1:\n", arreglo1)
print("\nArreglo 2:\n", arreglo2)

suma=arreglo1+arreglo2
print("\nSuma:\n", suma)

resta=arreglo1-arreglo2
print("\nResta:\n", resta)

multiplicacion=arreglo1*arreglo2
print("\nMultiplicación:\n", multiplicacion)
```

**P9.** Dado el arreglo `notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])`,  
usa **indexación booleana** para obtener solo las notas mayores o iguales a 6. Imprime el resultado.

```python
notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])

# Tu respuesta aquí
notas_mayores_igual_6=notas[notas>=6]
print("\nNotas mayores o iguales a 6:\n",notas_mayores_igual_6)
#
```
