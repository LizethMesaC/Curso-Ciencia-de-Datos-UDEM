**¿Cuántas personas tienen la palabra Chief en el título de su trabajo?** Apoyese en los métodos para manipular strings y en el método `.apply()`

```python
# Copie el código aca
print(salarios['JobTitle'].apply(lambda x: 'chief' in str(x).lower()).sum())
```

Este código es valido, se recomienda hacer el código mas simple y agregar typing y documentación.

```python
def contar_chiefs(df: pd.DataFrame) -> int:
    """

    Cuenta cuantas personas tiene la palabra 'Chief' en su título de trabajo ('JobTitle')

    """
    return df['JobTitle'].str.contains('chief', case=False, na=False).sum()
print(contar_chiefs(salarios))
```
